```python
3. What is the proportion of different types of content on Netflix?
```

```python
# Donut plot for content types
content_counts = dataframe['type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(content_counts, labels=content_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.gca().add_artist(plt.Circle((0, 0), 0.6, color='white'))  # Adding a white circle in the center
plt.title("Netflix Content Types Distribution")
plt.show()
```
